```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/loan_data.csv')
df.head()
```

```python
X = df.drop('loan_status', axis=1)
y = df['loan_status']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
print("MinMaxScaler imported successfully.")
```

```python
scaler = MinMaxScaler()
X_scaled_array = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled_array, columns=X.columns)
print("Min-Max scaling applied and scaled DataFrame created.")
X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression as LogR

# create the model
logR_algo = LogR(penalty='l2')
logR_algo
```

```python
logR_model = logR_algo.fit(X_train, y_train)
logR_model
```

```python
from sklearn.metrics import accuracy_score

# Predict the test data
predict = logR_model.predict(X_test)

# Separate the first five predictions and the first five real values in y_test
for i in range(5):
  print(f'Predicted: {round(predict[i],2)}')
  print(f'Real: {y_test.iloc[i]}') # Corrected to y_test and using .iloc for safe indexing
  print("\n")

print("\n")

print(f'Accuracy: {round(accuracy_score(y_test, predict),2)}')
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC

# create the model
DTC_algo = DTC()
DTC_algo
```

```python
DTC_model = DTC_algo.fit(X_train, y_train)
DTC_model
```

```python
from sklearn.metrics import accuracy_score

# Predict the test data
predict = DTC_model.predict(X_test)

# Separate the first five predictions and the first five real values in y_test
for i in range(5):
  print(f'Predicted: {round(predict[i],2)}')
  print(f'Real: {y_test.iloc[i]}') # Corrected to y_test and using .iloc for safe indexing
  print("\n")

print("\n")

print(f'Accuracy: {round(accuracy_score(y_test, predict),2)}')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay as CM
CM.from_predictions(y_test, predict)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# This is the list of parameters we will tune. Some are fixed values, some are distributions
tuned_parameters = {
    'criterion': ['gini', 'entropy'],
    'max_depth': randint(3, 9),  # Draw from a uniform distribution between 3 and 15
    'min_samples_split': randint(3, 9),  # Draw from a uniform distribution between 2 and 10
    'max_features': ['sqrt', 'log2', None]
}

# We will tune against f1_macro (to find the best choice)
score = 'f1_macro'

print(f"Tuning hyperparameters for {score}")
print("\n")

# Do the search using 5 folds/chunks
clf = RandomizedSearchCV(DTC(), tuned_parameters, cv=5, random_state=1984,
                         scoring=score, n_iter=20, refit=True)

# Pass the data to fit/train
clf.fit(X_train, y_train) # Using 'y_train' as per our dataset

print("Best parameters set found on the training set:")
print(clf.best_params_)
```

```python
DTC_algo_tuned = DTC(**clf.best_params_) # get the best parameters from randomized search
DTC_model_tuned = DTC_algo_tuned.fit(X_train, y_train) # Corrected Y_train to y_train

print("Decision Tree model retrained with optimized hyperparameters.")
```

```python
from sklearn.metrics import accuracy_score

# predict the test data using the tuned model
predict_tuned = DTC_model_tuned.predict(X_test)

# seperate the first five predictions and the first five real values in y_test
for i in range(5):
  print(f'Predicted: {round(predict_tuned[i],2)}')
  print(f'Real: {y_test.iloc[i]}') # Corrected to y_test and using .iloc for safe indexing
  print("\n")

print("\n")

print(f'Accuracy of Tuned DTC: {round(accuracy_score(y_test, predict_tuned),2)}')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay as CM, precision_recall_fscore_support
import matplotlib.pyplot as plt

# create a confusion matrix for the tuned model
cm_display_tuned = CM.from_predictions(y_test, predict_tuned)


# Calculate precision, recall, and F1-score for the tuned model
precision_micro_tuned, recall_micro_tuned, f1_score_micro_tuned, _ = precision_recall_fscore_support(y_test, predict_tuned, average='micro')

print(f"Micro-averaged Precision (Tuned DTC): {round(precision_micro_tuned, 2)}")
print(f"Micro-averaged Recall (Tuned DTC): {round(recall_micro_tuned, 2)}")
print(f"Micro-averaged F1-score (Tuned DTC): {round(f1_score_micro_tuned, 2)}")
print("\n")

precision_macro_tuned, recall_macro_tuned, f1_score_macro_tuned, _ = precision_recall_fscore_support(y_test, predict_tuned, average='macro')

print(f"Macro-averaged Precision (Tuned DTC): {round(precision_macro_tuned, 2)}")
print(f"Macro-averaged Recall (Tuned DTC): {round(recall_macro_tuned, 2)}")
print(f"Macro-averaged F1-score (Tuned DTC): {round(f1_score_macro_tuned, 2)}")
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15)) # Increased figure size for better readability
plot_tree(DTC_model_tuned, filled=True, feature_names=X_train.columns, class_names=['No Default', 'Default'], rounded=True)
plt.title('Tuned Decision Tree Visualization')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression as LogR
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.ensemble import GradientBoostingClassifier as GBDT
from xgboost import XGBClassifier as XGB
from sklearn.metrics import precision_recall_fscore_support

LogR_algo = LogR(random_state=42) # Added random_state for reproducibility
LogR_model = LogR_algo.fit(X_train, y_train)

RF_algo = RF(random_state=42) # Added random_state for reproducibility
RF_model = RF_algo.fit(X_train, y_train)

GBDT_algo = GBDT(random_state=42) # Added random_state for reproducibility
GBDT_model = GBDT_algo.fit(X_train, y_train)

XGB_algo = XGB(random_state=42, use_label_encoder=False, eval_metric='logloss') # Added random_state and suppressed warning for XGBoost
XGB_model = XGB_algo.fit(X_train, y_train)

models = [LogR_model, RF_model, GBDT_model, XGB_model]
names = ['Logistic Regression', 'Random Forest', 'GBDT', 'XGBoost'] # Corrected name for XGBoost

for i in range(len(models)):
  print(f"Model: {names[i]}")

  # predict based on training data
  predict = models[i].predict(X_train)

  # Calculate precision, recall, and F1-score
  precision, recall, f1_score, _ = precision_recall_fscore_support(y_train, predict, average='macro')
  print(f"Macro Precision: {round(precision, 4)}") # Added rounding for better readability
  print(f"Macro Recall: {round(recall, 4)}") # Added rounding for better readability
  print(f"Macro F1-score: {round(f1_score, 4)}") # Added rounding for better readability
  print("\n")
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

# we get a load of warnings running the code so will supress them
import warnings
warnings.filterwarnings("ignore")

# create a hyperparameter search function for re-usability
def random_search(algo, hyperparameters, X_train, y_train):
  # do the search using 5 folds/chunks
  clf = RandomizedSearchCV(algo, hyperparameters, cv=5, random_state=2015,
                          scoring='precision_macro', n_iter=20, refit=True)

  # pass the data to fit/train
  clf.fit(X_train, y_train)

  return clf.best_params_


# Random Forest
RF_tuned_parameters = {
    'n_estimators': randint(50, 500), # Draw from a uniform distribution between 50 and 500
    'max_depth': randint(2, 7),  # Draw from a uniform distribution between 2 and 7
    'min_samples_split': randint(2, 7),  # Draw from a uniform distribution between 2 and 7
    'max_features': ['sqrt', 'log2', None]
}

RF_best_params = random_search(RF_algo, RF_tuned_parameters, X_train, y_train)

# Train the models

RF_algo = RF(**RF_best_params)
RF_model = RF_algo.fit(X_train, y_train)

# score the models
models = [RF_model] # redo this or it uses the old models

for i in range(len(models)): # for every model in the models list
  print(f"Model: {'Random forest'}")

  # predict based on training data
  predict = models[i].predict(X_train)

  # Calculate precision, recall, and F1-score
  precision, recall, f1_score, _ = precision_recall_fscore_support(y_train, predict, average='macro')
  print(f"Macro Precision: {precision}")
  print(f"Macro Recall: {recall}")
  print(f"Macro F1-score: {f1_score}")
  print("\n")
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, ConfusionMatrixDisplay as CM
import matplotlib.pyplot as plt

# Predict the test data using the Random Forest model
rf_predict = RF_model.predict(X_test)

# Calculate accuracy
rf_accuracy = accuracy_score(y_test, rf_predict)
print(f'Random Forest Model Accuracy: {round(rf_accuracy, 4)}')

# Calculate precision, recall, and F1-score (macro-averaged)
rf_precision, rf_recall, rf_f1_score, _ = precision_recall_fscore_support(y_test, rf_predict, average='macro')
print(f'Random Forest Model Macro Precision: {round(rf_precision, 4)}')
print(f'Random Forest Model Macro Recall: {round(rf_recall, 4)}')
print(f'Random Forest Model Macro F1-score: {round(rf_f1_score, 4)}')

# Display Confusion Matrix
print('\nConfusion Matrix for Random Forest Model:')
cm_display_rf = CM.from_predictions(y_test, rf_predict)
plt.title('Confusion Matrix for Random Forest Model')
plt.show()
```
